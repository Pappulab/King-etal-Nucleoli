```python
import pandas as pd
import numpy as np

# Read in localization excel file 
locdata=pd.read_excel('../Bioinformatics_Data/UMAP/umap_results_fit_all_transform_all_sorted_20190422.xlsx') 

print(locdata)
```

```python
# Read in Mapping table
mapdata=pd.read_excel('../Bioinformatics_Data/UMAP/ENSEMBL_to_Uniprot_location_proteins.xlsx') 

print(mapdata)
```

```python
# Add uniprot ID to dataframe
ensembl_list_1 = locdata['gene_ensembl_id'].tolist()
ensembl_list_2 = mapdata['ENSEMBL'].tolist()
uni_list = mapdata['Uniprot'].tolist()

allunimatch=[]
for ens in ensembl_list_1:
    if ens in ensembl_list_2:
        curridx = [i for i, x in enumerate(ensembl_list_2) if x == ens]
        tmp=" ".join(uni_list[x] for x in curridx)
        allunimatch.append(tmp)
    else:
        allunimatch.append('')
        
print(allunimatch)
```

```python
locdata['Uniprot']=allunimatch
print(locdata)
#locdata.to_excel('../Localization_Data/umap_results_fit_all_transform_all_sorted_20190422_with_uniprot.xlsx')
```
